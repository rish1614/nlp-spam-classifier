## Imports and NLTK Setups

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

## Presprocessing

```python
df = pd.read_csv('../data/raw/spam.tsv', sep='\t', header=None, names=['label', 'message'])
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    # 1. Lowercase
    text = text.lower()
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # 3. Remove punctuation and numbers
    text = re.sub(r'[^a-z\s]', '', text)
    # 4. Tokenize and remove stopwords/lemmatize
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)

# Apply cleaning
df['clean_message'] = df['message'].apply(clean_text)

# Drop empty messages post-cleaning
df = df[df['clean_message'].str.strip() != '']

# Save processed data
df.to_csv('../data/processed/cleaned_spam.csv', index=False)
print("Preprocessing complete. Data saved to ../data/processed/")
```
