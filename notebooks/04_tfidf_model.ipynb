## Setup and TF-IDF

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report

df = pd.read_csv('../data/processed/cleaned_spam.csv').dropna()
X_train, X_test, y_train, y_test = train_test_split(df['clean_message'], df['label'], test_size=0.2, random_state=42, stratify=df['label'])

# TF-IDF Vectorizer
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

## Training Logistic Regression and SVM

```python
print("--- Logistic Regression ---")
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_tfidf, y_train)
print(classification_report(y_test, lr.predict(X_test_tfidf)))

print("--- Support Vector Machine (Linear) ---")
svm = SVC(kernel='linear', probability=True)
svm.fit(X_train_tfidf, y_train)
print(classification_report(y_test, svm.predict(X_test_tfidf)))

# SVM usually performs incredibly well on TF-IDF text data
```
